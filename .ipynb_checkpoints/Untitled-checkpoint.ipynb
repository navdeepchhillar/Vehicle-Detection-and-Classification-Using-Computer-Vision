```python
import cv2
import time
import uuid
from collections import defaultdict
import numpy as np
from sort import Sort 
import torch
import matplotlib.pyplot as plt
```

```python
# Load YOLOv5 model (yolov5m in this case)
model = torch.hub.load('ultralytics/yolov5', 'custom', path='C:\\Users\\Balaji\\Vehicle-Detection-and-Classification-Using-Computer-Vision\\yolov5m.pt', force_reload=True)
```

```python
# Function to process the video
def process_video(video_path, polygon_points, conf_threshold=0.5):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video {video_path}")
        return

    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    frame_size = (width, height)

    output_path = 'C:\\Users\\Balaji\\Vehicle-Detection-and-Classification-Using-Computer-Vision\\output.mp4'
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, frame_size)
    if not out.isOpened():
        print("Error: Could not open VideoWriter.")
        return

    vehicle_count = {"car": 0, "bus": 0, "truck": 0}
    confidence_per_frame = []
    tracked_objects = {}
    vehicle_ids_in_polygon = set()

    def draw_polygon(frame, polygon_points):
        return cv2.polylines(frame, [polygon_points], isClosed=True, color=(0, 255, 255), thickness=2)

    def is_in_polygon(point, polygon):
        return cv2.pointPolygonTest(polygon, point, False) >= 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame)
        frame_confidences = []
        detections = []

        for detection in results.xyxy[0]:
            x1, y1, x2, y2, conf, cls = detection
            if conf < conf_threshold:
                continue

            label = model.names[int(cls)]
            if label not in vehicle_count:
                continue

            x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))
            center = (int((x1 + x2) / 2), int((y1 + y2) / 2))

            for polygon in polygon_points:
                if is_in_polygon(center, polygon):
                    obj_id = None

                    # Try to match an existing object
                    for tid, data in tracked_objects.items():
                        tx1, ty1, tx2, ty2 = data["bbox"]
                        tcenter = ((tx1 + tx2) // 2, (ty1 + ty2) // 2)
                        if abs(tcenter[0] - center[0]) < 30 and abs(tcenter[1] - center[1]) < 30:
                            obj_id = tid
                            break

                    # If new object, create ID
                    if obj_id is None:
                        obj_id = str(uuid.uuid4())
                        tracked_objects[obj_id] = {"bbox": (x1, y1, x2, y2), "label": label}
                        if obj_id not in vehicle_ids_in_polygon:
                            vehicle_count[label] += 1
                            vehicle_ids_in_polygon.add(obj_id)
                    else:
                        tracked_objects[obj_id]["bbox"] = (x1, y1, x2, y2)

                    # Draw box and label
                    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                    cv2.putText(frame, f"{label}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2)

            frame_confidences.append(conf.item())

        avg_conf = sum(frame_confidences) / len(frame_confidences) if frame_confidences else 0
        confidence_per_frame.append(avg_conf)

        # Draw polygons
        for polygon in polygon_points:
            draw_polygon(frame, polygon)

        # Show count
        count_text = f"Cars: {vehicle_count['car']} | Buses: {vehicle_count['bus']} | Trucks: {vehicle_count['truck']}"
        cv2.putText(frame, count_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

        out.write(frame)
        cv2.imshow('Vehicle Detection', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    out.release()
    cv2.destroyAllWindows()

    print("Video processing complete. Output saved at:", output_path)

    plt.figure(figsize=(12, 5))
    plt.plot(confidence_per_frame, 'bo-', markersize=3)
    plt.title('Average Detection Confidence Per Frame')
    plt.xlabel('Frame Number')
    plt.ylabel('Average Confidence')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    # Define the video path and polygon coordinates
video_path = 'C:\\Users\\Balaji\\Vehicle-Detection-and-Classification-Using-Computer-Vision\\56310-479197605.mp4'
polygon_points = [
    np.array([(1040, 459), (1362, 441), (1477, 515), (1081, 555)], np.int32),
    np.array([(785, 471), (406, 434), (279, 503), (746, 574)], np.int32)
]

# Call the function to process the video
process_video(video_path, polygon_points, conf_threshold=0.5)


#Navdeep this side for any problem contact 7982447979
```
